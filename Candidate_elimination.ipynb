```python
import csv

# Read the CSV file
with open('enjoy_sport.csv', 'r') as file:
    data = list(csv.reader(file))

# Separate attributes and target class
concepts = [row[:-1] for row in data[1:]]
target = [row[-1] for row in data[1:]]

# Number of attributes
num_attributes = len(concepts[0])

# Initialize Specific Hypothesis
S = concepts[0].copy()

# Initialize General Hypothesis
G = [['?' for _ in range(num_attributes)] for _ in range(num_attributes)]

print("Initial Specific Hypothesis:")
print(S)

print("\nInitial General Hypothesis:")
for row in G:
    print(row)

# Candidate Elimination Algorithm
for i in range(len(concepts)):

    if target[i].lower() == 'yes':

        # Update Specific Hypothesis
        for j in range(num_attributes):
            if concepts[i][j] != S[j]:
                S[j] = '?'
                G[j][j] = '?'

    else:

        # Update General Hypothesis for negative example
        for j in range(num_attributes):
            if concepts[i][j] != S[j]:
                G[j][j] = S[j]
            else:
                G[j][j] = '?'

    print(f"\nAfter Example {i + 1}:")
    print("Specific Hypothesis:")
    print(S)

    print("General Hypothesis:")
    for row in G:
        print(row)

# Remove rows that contain only '?'
final_G = []

for row in G:
    if row != ['?' for _ in range(num_attributes)]:
        final_G.append(row)

print("\nFinal Specific Hypothesis:")
print(S)

print("\nFinal General Hypothesis:")
for row in final_G:
    print(row)
```
